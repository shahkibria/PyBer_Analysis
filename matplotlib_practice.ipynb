```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot and add a label, colour and line width
plt.subplots(figsize=(10, 5))
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=1, label='Boston')
# Create a title
plt.title("Pyber Fare by Month")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Add a grid
plt.grid()
# Set the y limit between 5 and 45.
plt.ylim(5, 45)
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, label = "Boston", marker="*", color="green", linewidth=1)
# Add X and Y Axis labels
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Add a grid
ax.grid()
# Add a legend
ax.legend()
# Set a Title
ax.set_title("Pyber Fare by Month")
# Set the y limit between 5 and 45.
ax.set_ylim(5,45)
```

```python
# Create a bar Chart
plt.subplots(figsize=(6,6))
plt.bar(x_axis, y_axis, label="Boston", color = "red")
# Create a title
plt.title("Pyber Fare by Month")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Add the legend.
plt.legend()
```

```python
# Create a horizontal bar chart. Note = for a regular vertical bar chart, just replace
# the "h" and the code will be "plt.bar"
plt.subplots(figsize=(10, 10))
plt.barh(x_axis, y_axis, label="Boston", color = "black")
# Create a title
plt.title("Pyber Fare by Month")
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Add the legend.
plt.legend()
```

```python
# Create the bar plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, label = "Boston", color="Black")
# Add X and Y Axis labels
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Add a legend
ax.legend()
# Set a Title
ax.set_title("Pyber Fare by Month")
```

```python
plt.plot(x_axis, y_axis, 'o',label="Boston", color = "black")
plt.title("Pyber Fare by Month")
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Add the legend.
plt.legend()
```

```python
plt.scatter(y_axis, x_axis,label = "Chicago", color = "black")
plt.title("Pyber Fare by Month")
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Add the legend.
plt.legend()
#Reverse the Y axis
plt.gca().invert_yaxis()
```

```python
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*5)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s=[i*5 for i in y_axis])
```

```python
# Create the bar plot with ax.plt()
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, label = "Chicago", color="Black",s=[i*5 for i in y_axis])
# Add the legend.
plt.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0.6, 0, 0, 0, 0, 0, .3, 0, 0, 0, 0, 0)
plt.subplots(figsize=(5, 20))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```
