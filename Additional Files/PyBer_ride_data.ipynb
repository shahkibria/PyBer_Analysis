```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Import CSV File
import csv
import os
pyber_data_to_load = os.path.join("Resources", "PyBer_ride_data.csv")
pyber_ride_df = pd.read_csv(pyber_data_to_load)
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", marker="*", color="blue", 
                   linewidth=1, label='Boston',figsize=(15, 5))
# Add a grid
plt.grid()
plt.ylabel("Fare($)")
pyber_ride_df
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)",marker="*", color="blue", 
                   linewidth=1, label='Boston',figsize=(10, 5))
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.grid()
plt.ylabel("Fare($)")
# Set the y limit between 5 and 45.
plt.xlim(0, 11)
plt.show()
```

```python
# Plot the Bar Chart data.
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)", color="blue", 
                label='Boston',figsize=(5, 7))
```
